```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
arr = tf.constant([-9, -8.5, -4.32, -1, -0, 0, 2.89, 5.76])

tf.keras.activations.relu(arr)
```

```python
plt.plot(arr,tf.keras.activations.relu(arr))
```
